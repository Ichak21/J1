# Analyse des valeurs de donn√©es

```python
print(wine_dataset["quality"].describe())

dist_red_quality = wine_dataset[wine_dataset['type'] == "red" ]["quality"]
sns.histplot(dist_red_quality, legend=True, color='red')

dist_white_quality = wine_dataset[wine_dataset['type'] == "white" ]["quality"]
sns.histplot(dist_white_quality, legend=True, color="white")
```

```python
for col in wine_dataset.select_dtypes('float'):
    for quality in wine_dataset["quality"].value_counts().index.sort_values():
        print(" ============= > Analyse de " + str(col) + " pour la qualit√© " + str(quality))
        plt.figure()
        sns.histplot(wine_dataset[(wine_dataset["type"]=="red") & (wine_dataset["quality"]==quality)][col], kde=True, color='red')
        sns.histplot(wine_dataset[(wine_dataset["type"]=="white") & (wine_dataset["quality"]==quality)][col], kde=True, color='white')
        plt.show()
```
